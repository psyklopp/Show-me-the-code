### ðŸ“‰ Cost Function for Regression â€“ Mean Squared Error (MSE)

The **Mean Squared Error (MSE)** is a commonly used cost function in regression tasks. It measures the average of the squares of the errors â€” that is, the average squared difference between the predicted and actual values.

---

#### **Prediction Formula:**

In simple linear regression, predictions are made using:

$$
\hat{y} = w \cdot x + b
$$

Where:
- $x$ = input feature  
- $w$ = weight (slope of the line)  
- $b$ = bias (intercept)  
- $\hat{y}$ = predicted output  

---

#### **Cost Function Formula:**

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} \left( \hat{y}^{(i)} - y^{(i)} \right)^2
$$

Where:
- $n$ = number of data points  
- $\hat{y}^{(i)}$ = predicted value for the $i$-th data point  
- $y^{(i)}$ = actual (true) value for the $i$-th data point  

---

#### **Explanation:**
- The error $\hat{y}^{(i)} - y^{(i)}$ is the difference between the predicted and actual value.
- Squaring the error ensures that negative and positive errors donâ€™t cancel out.
- Averaging the squared errors gives the overall "cost" the model is paying for making incorrect predictions.

---

#### **Goal:**
Minimize the MSE during training to make the predictions as close as possible to the actual target values.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sample training data (simple linear relation: y = 2x)
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Create a grid of weight (w) and bias (b) values
w_vals = np.linspace(0, 4, 100)
b_vals = np.linspace(-2, 2, 100)
W, B = np.meshgrid(w_vals, b_vals)

# Function to compute Mean Squared Error
def compute_mse(w, b, X, y):
    m = len(X)
    predictions = w * X[:, np.newaxis, np.newaxis] + b
    errors = predictions - y[:, np.newaxis, np.newaxis]
    mse = np.mean(errors ** 2, axis=0)
    return mse

# Compute MSE for each (w, b) pair
Z = compute_mse(W, B, X, y)

# Create contour plot
fig, ax = plt.subplots(figsize=(8, 6))
contour = ax.contourf(W, B, Z, levels=50, cmap='viridis')
cbar = plt.colorbar(contour)
cbar.set_label('Cost (MSE)')

# Axis labels and title
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_title('Contour Plot of Cost Function (MSE)')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Plotting the cost function
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
surf = ax.plot_surface(W, B, Z, cmap='viridis', edgecolor='none')

# Axis labels
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Cost (MSE)')
ax.set_title('3D Cost Function (MSE) Surface')

# Show plot
plt.tight_layout()
plt.show()
```
